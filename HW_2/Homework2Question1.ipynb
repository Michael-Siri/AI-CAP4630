# This code uses three_dimensional_plotting.ipynb and density_and_contour_plots.ipynb to display contour, density and plot of 3 simple bivariate functions.

A minimum of function $x^2 + y^2$

A maximum of function $-x^2 - y^2$

And one with neither of function $-x^2 + y^2$

```
#Michael Siripongpibul
#CAP-4630 Spring 2020

from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# Minimum

```
def f(x, y):
    return x ** 2 + y ** 2

x = np.linspace(-6, 6, 50)
y = np.linspace(-6, 6, 50)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter(x[25], y[25], alpha=1, color='r', s=200)
```

```
#Contour plot
plt.contour(X,Y,Z, 50)
plt.scatter(0,0)
plt.show()
```

```
#Density Plot
plt.contourf(X,Y,Z,50)
plt.show()
```

# Maximum

```
def d(x, y):
    return -x ** 2 - y ** 2

xx = np.linspace(-6, 6, 50)
yy = np.linspace(-6, 6, 50)

XX, YY = np.meshgrid(xx, yy)
ZZ = d(XX, YY)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(XX, YY, ZZ, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter(xx[25], yy[25], alpha=1, color='r', s=200)
```

```
#Contour plot
plt.contour(XX,YY,ZZ, 50)
plt.scatter(0,0)
plt.show()
```

```
#Density Plot
plt.contourf(XX,YY,ZZ,50)
plt.show()
```

# Neither

```
def g(x, y):
    return -x ** 2 + y ** 2

xxx = np.linspace(-6, 6, 50)
yyy = np.linspace(-6, 6, 50)

XXX, YYY = np.meshgrid(xxx, yyy)
ZZZ = g(XXX, YYY)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(XXX, YYY, ZZZ, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter(xxx[25], yyy[25], alpha=1, color='r', s=200)
```

```
#Contour plot
plt.contour(XXX,YYY,ZZZ, 50)
plt.scatter(0,0)
plt.show()
```

```
#Density Plot
plt.contourf(XXX,YYY,ZZZ,50)
plt.show()
```
